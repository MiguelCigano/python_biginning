Señales continuas periódicas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = 1
x = lambda t:np.sin(w*t)
t0 = -2*np.pi
tf = 6*np.pi
n_muestras = 401
```

```python
t = np.linspace(t0, tf, n_muestras)
```

```python
signal = x(t)
```

```python
plt.plot(t, signal, color= "red")
plt.axhline(0, color = "black")
plt.axvline(0, color="blue")
plt.xlabel("Tiempo (t)")
plt.ylabel("Señal x(t)")
plt.grid(True)
```

```python
np.set_printoptions(precision = 5)
print("Señal: ")
print(signal)
print(signal[0])
```

## Suma de señales periodicas

Sea $x_{1} (t) = \cos (\frac{\pi t}{2})$ y $x_{2}(t) = \cos (\frac{\pi t}{3})$, periódicas con periodos fundamentales de $T_{1} = 4$ y $T_{2} = 6$, respectivamente:

$$\frac{T_{1}}{T_{2}} = \frac{4}{6}$$,

Entonces, con $m = 2$ y $r =3$ se desprende que la suma $x_{1}(t) + x_{2}(t)$ es periódica con un periodo fundamental $rT_{1} = (3)(4) = 12 seg$, es decir:

$$\frac{r T_{1}}{m T_{2}} = \frac{3 \times 4}{2 \times 6} = 12 $$,

```python
w1 = np.pi / 2
w2 = np.pi / 3
x1 = lambda t:np.cos(w1*t)
x2 = lambda t:np.cos(w2*t)
t01 = -2*np.pi
t02 = -2*np.pi
tf1 = 6*np.pi
tf2 = 6*np.pi
n_muestras1 = 1001
n_muestras2 = 1001
```

```python
t1 = np.linspace(t01, tf1, n_muestras1)
t2 = np.linspace(t01, tf2, n_muestras2)
```

```python
signal1 = x1(t1)
signal2 = x2(t2)
signal3 = x1(t1) + x2(t2)
```

```python
plt.plot(t1, signal1, color= "magenta")
plt.plot(t2, signal2, color= "red")
plt.axhline(0, color = "black")
plt.axvline(0, color="blue")
plt.xlabel("Tiempo (t)")
plt.ylabel("Señal x(t)")
plt.grid(True)
```

```python
plt.plot(t1, signal3, color= "blue")
plt.axhline(0, color = "black")
plt.axvline(0, color="blue")
plt.xlabel("Tiempo (t)")
plt.ylabel("Señal x(t)")
plt.grid(True)
```
