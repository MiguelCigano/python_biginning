### Uso del paquete random para generación de número aleatorios



```python
import random
```

Un número aleatorio que va desde el cero hasta el $1$, pero sin incluir el $1$, es decir, [0, 1).

image.png

```python
random.random()
```

```python
random.random()
```

Podemos especificar un rango en el que queremos el número aleatorio a través de le método **uniform**. Considera que estos métodos son abiertos por la izquierda, para el ejemplo: $[0, n)$, donde $n = 10.$

```python
random.uniform(1, 10)
```

```python
random.uniform(1, 10)
```

Si queremos un rango de número enteros pero no flotantes, podemos usar el método **randrange**. Recuerda que no incluye el número por la derecha.

```python
random.randrange(10)
```

```python
random.randrange(10)
```

El metodo **randrange** nos permite además especificar un múltiplo por el cual se quiere obtener un número aleatorio.

```python
random.randrange(0, 10, 3)
```

Además, el paquete random no está limitado a números.

```python
cadena = "Universidad_de_Tijuana"
random.choice(cadena)
```

```python
lista = [1, 2, 3, cadena, 3.1416]
random.choice(lista)
```

```python
random.shuffle(lista)
lista
```

```python
random.sample(lista, 3)
```
